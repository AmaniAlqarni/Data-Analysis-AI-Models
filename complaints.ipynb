```python
# الحالات المصنفة كمغلقة
closed_statuses = [
    'إغلاق لصالح المستفيد',
    'إغلاق لصالح مقدم الخدمة',
    'توجيه بالمعالجة',
    'تمت المعالجة',
    'مقبول',
    'مرفوض',
    'مكتمل'
]

# الحالات المصنفة كقيد المعالجة
in_progress_statuses = [
    'قيد الانتظار (موافقة مدير الشكاوى)',
    'قيد الانتظار (موافقة لجنة التدقيق)',
    'قيد المراجعة',
    'قيد المراجعة (متأخر)'
]

# إنشاء العمود الموحد بناءً على الحالتين
df_allcomplaints['الحالة الموحدة'] = df_allcomplaints['حالة الشكوى'].apply(
    lambda x: 'مغلقة' if x in closed_statuses
    else 'قيد المعالجة' if x in in_progress_statuses
    else 'أخرى'
)


#display(df_allcomplaints[['حالة الشكوى', 'الحالة الموحدة']])


new_status = df_allcomplaints['الحالة الموحدة'].value_counts().reset_index()
display(new_status["count"].sum())
display(new_status)
```

```python
total_closed_status = df_allcomplaints['حالة الشكوى'].value_counts().reset_index()
display(total_closed_status["count"].sum())
display(total_closed_status)
```

```python
# Closure Rate
total_complaints = df_allcomplaints.shape[0]
closed_count = df_allcomplaints[df_allcomplaints['الحالة الموحدة'] == 'مغلقة'].shape[0]
closure_rate = round((closed_count / total_complaints) * 100, 2) if total_complaints else 0

print(" عدد الشكاوى:", total_complaints)
print(" عدد الشكاوى المغلقة:", closed_count)
print(" معدل الإغلاق:", closure_rate, "%")
```

```python
# Average of Colsure Rate
avg_days = df_allcomplaints['المدة الفعلية (أيام العمل)'].mean().round(2)
print(avg_days)
```

```python
# Complaints Volume By Status
status = df_allcomplaints["حالة الشكوى"].value_counts().reset_index()
display(status["count"].sum())
display(status)
```

```python
# Closed Complaints Based on Decision and Duration
# Targeted Status
target_statuses = ['إغلاق لصالح المستفيد', 'إغلاق لصالح مقدم الخدمة', 'توجيه بالمعالجة', 'تمت المعالجة']
total_complaints = df_allcomplaints.shape[0]
results = []

for status in target_statuses:
    # On-Time Complaints
    on_time = df_allcomplaints[
        (df_allcomplaints['حالة الشكوى'] == status) &
        (df_allcomplaints['المدة الفعلية (أيام العمل)'] <= 5)
    ]
    on_time_avg = round(on_time['المدة الفعلية (أيام العمل)'].mean(), 2)
    on_time_count = on_time.shape[0]
    on_time_percent = round((on_time_count / total_complaints) * 100, 2)

    results.append({
        'حالة الشكوى': status,
        'نوع المعالجة': 'في الوقت',
        'عدد الشكاوى': on_time_count,
        'متوسط المدة': on_time_avg,
        'النسبة المئوية من الإجمالي': on_time_percent
    })

    # Late Complaints
    late = df_allcomplaints[
        (df_allcomplaints['حالة الشكوى'] == status) &
        (df_allcomplaints['المدة الفعلية (أيام العمل)'] > 5)
    ]
    late_avg = round(late['المدة الفعلية (أيام العمل)'].mean(), 2)
    late_count = late.shape[0]
    late_percent = round((late_count / total_complaints) * 100, 2)

    results.append({
        'حالة الشكوى': status,
        'نوع المعالجة': 'متأخرة',
        'عدد الشكاوى': late_count,
        'متوسط المدة': late_avg,
        'النسبة المئوية من الإجمالي': late_percent
    })


summary_df = pd.DataFrame(results)
display(summary_df)
```

```python
# Monthly Escalated Complaints Based On Region
df_allcomplaints['تاريخ التصعيد'] = pd.to_datetime(df_allcomplaints['تاريخ التصعيد'], errors='coerce')
df_allcomplaints['شهر التصعيد'] = df_allcomplaints['تاريخ التصعيد'].dt.to_period('M').astype(str)
monthly_city_counts = df_allcomplaints.groupby(['شهر التصعيد', 'المنطقة']).size().reset_index(name='عدد الشكاوى')


top3_and_others = []
for month, group in monthly_city_counts.groupby('شهر التصعيد'):
    top3 = group.sort_values('عدد الشكاوى', ascending=False).head(3)
    others = group.sort_values('عدد الشكاوى', ascending=False).iloc[3:]
    others_total = others['عدد الشكاوى'].sum()

    for _, row in top3.iterrows():
        top3_and_others.append({
            'الشهر': month,
            'المنطقة': row['المنطقة'],
            'عدد الشكاوى': row['عدد الشكاوى']
        })

    top3_and_others.append({
        'الشهر': month,
        'المنطقة': 'بقية المناطق',
        'عدد الشكاوى': others_total
    })

result = pd.DataFrame(top3_and_others)
display(result)
```

```python
# Monthly Escalated Complaints Based On Complaints Type
df_allcomplaints['تصنيف الشكوى'] = df_allcomplaints['تصنيف الشكوى'].replace('-', 'غير مصنف')
df_allcomplaints['تصنيف الشكوى'] = df_allcomplaints['تصنيف الشكوى'].fillna('غير مصنف')
df_allcomplaints['تاريخ التصعيد'] = pd.to_datetime(df_allcomplaints['تاريخ التصعيد'], errors='coerce')
df_allcomplaints['شهر التصعيد'] = df_allcomplaints['تاريخ التصعيد'].dt.to_period('M').astype(str)
grouped_by_issue = df_allcomplaints.groupby(['شهر التصعيد', 'تصنيف الشكوى']).size().reset_index(name='عدد الشكاوى')
grouped_by_issue = grouped_by_issue.sort_values('شهر التصعيد', ascending=False)
display(grouped_by_issue)
```

```python
# Monthly Escalated Complaints Based On Volume, CLosure Rate and Duration
df_allcomplaints['تاريخ التصعيد'] = pd.to_datetime(df_allcomplaints['تاريخ التصعيد'], errors='coerce')
df_allcomplaints['تاريخ اغلاق الشكوى'] = pd.to_datetime(df_allcomplaints['تاريخ اغلاق الشكوى'], errors='coerce')
df_allcomplaints['شهر التصعيد'] = df_allcomplaints['تاريخ التصعيد'].dt.to_period('M')

# الحالات الموحدة
status_categories = ['مغلقة', 'قيد المعالجة']
monthly_volume = df_allcomplaints[df_allcomplaints['تاريخ التصعيد'].notnull()].groupby('شهر التصعيد').agg(
    اجمالي_الشكاوى=('رقم الشكوى', 'count'),
    عدد_المغلقة=('تاريخ اغلاق الشكوى', lambda x: x.notnull().sum()),
    متوسط_المدة=('المدة الفعلية (أيام العمل)', 'mean')
).reset_index()

for status in status_categories:
    count_by_status = df_allcomplaints[df_allcomplaints['الحالة الموحدة'] == status].groupby('شهر التصعيد').size()
    monthly_volume[f'عدد - {status}'] = monthly_volume['شهر التصعيد'].map(count_by_status).fillna(0).astype(int)

# "إضافة عدد "توجيه بالمعالجة
guidance_counts = df_allcomplaints[df_allcomplaints['حالة الشكوى'] == 'توجيه بالمعالجة'].groupby('شهر التصعيد').size()
monthly_volume['عدد - توجيه بالمعالجة'] = monthly_volume['شهر التصعيد'].map(guidance_counts).fillna(0).astype(int)

# حساب معدل الإغلاق
monthly_volume['معدل الإغلاق'] = (monthly_volume['عدد_المغلقة'] / monthly_volume['اجمالي_الشكاوى'] * 100).round(2)
monthly_volume['متوسط_المدة'] = monthly_volume['متوسط_المدة'].round(1)
monthly_volume['شهر التصعيد'] = monthly_volume['شهر التصعيد'].astype(str)
monthly_volume = monthly_volume.sort_values('شهر التصعيد', ascending=False)
display(monthly_volume)
```
