## City

```python
# Remove leading/trailing whitespaces from 'المدينة'
df_complaints_SocialMedia["المدينة"] = df_complaints_SocialMedia["المدينة"].str.strip()
```

```python
# Remove leading/trailing whitespaces from 'المدينة'
df_complaints_SocialMedia["المدينة"] = df_complaints_SocialMedia["المدينة"].str.strip()

# Adjust some Cities name
df_complaints_SocialMedia["المدينة"]= df_complaints_SocialMedia["المدينة"].map({"الخُبر":"الخبر",
                                                                                                     "إيصال":"الدمام",
                                                                                                      "الهفوف":"الأحساء",
                                                                                                       "حازان" : "جازان",
                                                                                                       "الاحساء" : "الأحساء",
                                                                                                       "جازان (صبيا)" : "جازان",
                                                                                                       "ابها" : "أبها",
                                                                                                        "مكة" : "مكة المكرمة" ,
                                                                                                        "حريملا" : "حريملاء"}).fillna(df_complaints_SocialMedia["المدينة"])
# Define a function to map values
def assign_value(المدينة):
    if pd.isna(المدينة):
        return np.nan  # or return city, depending on desired behavior
    elif المدينة in ('الرياض', 'الخرج', "الدرعية", "الدوادمي","الزلفي","مرات","شقراء","ساجر", "ثادق", "حريملاء","سدوس","سدير", "وادي الدواسر", "عفيف", "الحريق", "المجمعة", "المزاحمية"):
        return 'الرياض'
    
    elif المدينة in ('الباحة', 'بلجرشي', "المندق", "القرى", "المخواه"):
        return "الباحة"
    
    elif المدينة in ('سكاكا', 'الجوف', "طبرجل", "دومة الجندل", "القريات"):
        return "الجوف"
    
    elif المدينة in ('عرعر', "طريف", "الحدود الشمالية", "الحدود الشماليه", "رفحاء"):
        return "الحدود الشمالية"
    
    elif المدينة in ('الدمام', 'الأحساء', "الخبر", "الظهران","الجبيل","القطيف","حفر الباطن","سيهات", "الخفجي", "الاحساء","القيصومة", "الشرقية", "بقيق"):
        return "الشرقية"
    
    elif المدينة in ('بريدة', 'القصيم', "عنيزة", "بريده", "الاسياح", "الأسياح", "الرس", "المذنب"):
        return "القصيم"
    
    elif المدينة in ('المدينة المنورة', 'ينبع', "العلا", "المهد"):
        return "المدينة المنورة"
    
    elif المدينة in ('تبوك', 'أملج', "الوجه", "نيوم", "ضباء", "حقل", ):
        return "تبوك"
    
    elif المدينة in ('جازان', 'أبو عريش', "صامطة", "جازان (صبيا)", "جيزان", "شقيق", "الشقيق", "صبيا", "الداير"):
        return "جازان"
    
    elif المدينة in ('حائل'):
        return "حائل"
    
    elif المدينة in ('أبها', 'عسير', "بيشة", "أحد رفيدة", "ابها", 'خميس مشيط',"النماص", "تثليث", "بلقرن", "رجال ألمع"):
        return "عسير"
    
    elif المدينة in ('جدة', 'مكة المكرمة', "الطائف", "القنفذة","رابغ", "الهدا", "جده", "الليث", "الخرمة"):
        return "مكة المكرمة"
    
    elif المدينة in ('نجران', "شرورة", "حبونا"):
        return "نجران"
    
    else:
        return 'Check'

# Apply the function to create the new column
df_complaints_SocialMedia['المنطقة'] = df_complaints_SocialMedia['المدينة'].apply(assign_value)   
                                                                                                 

Region_temp = df_complaints_SocialMedia["المنطقة"].value_counts(dropna=False).reset_index()
Region_temp_no_nulls = df_complaints_SocialMedia["المنطقة"].value_counts().reset_index()

print("With Nulls", Region_temp["count"].sum())
print("without Nulls",Region_temp_no_nulls["count"].sum())
display(Region_temp)
City_temp = df_complaints_SocialMedia["المدينة"].value_counts(dropna=False).reset_index()
display(City_temp)
```
