```python
# import pandas as pd
# from math import radians, sin, cos, sqrt, atan2

# def haversine(lat1, lon1, lat2, lon2):
#     """
#     Calculates the Haversine distance between two points on Earth.

#     Args:
#         lat1 (float): Latitude of the first point.
#         lon1 (float): Longitude of the first point.
#         lat2 (float): Latitude of the second point.
#         lon2 (float): Longitude of the second point.

#     Returns:
#         float: The distance in kilometers.
#     """
#     R = 6371  # Radius of Earth in kilometers

#     lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

#     dlon = lon2 - lon1
#     dlat = lat2 - lat1

#     a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
#     c = 2 * atan2(sqrt(a), sqrt(1 - a))

#     distance = R * c
#     return distance


# def add_closest_district_info_to_target_df(districts, target_df):
#     """
#     Adds the 'name', 'اسم المحافظة باللغة العربية', and 'اسم المنطقة باللغة العربية'
#     columns of the closest district to each row in the target DataFrame.

#     Args:
#         districts (pd.DataFrame): DataFrame with district information
#                                   ('name', 'longitude', 'latitude',
#                                    'اسم المحافظة باللغة العربية',
#                                    'اسم المنطقة باللغة العربية').
#         target_df (pd.DataFrame): DataFrame to which the closest district
#                                  information will be added ('longitude', 'latitude').

#     Returns:
#         pd.DataFrame: The target_df with the added columns.
#     """

#     def get_closest_district_info(target_lon, target_lat):
#         """Helper function to find the closest district for a single target coordinate
#            and return the specified columns.
#         """
#         distances = districts.apply(
#             lambda row: haversine(target_lat, target_lon, row['latitude'], row['longitude']),
#             axis=1
#         )
#         closest_district_index = distances.idxmin()  # Index of the minimum distance
#         closest_district = districts.loc[closest_district_index]
#         return closest_district[['name', 'اسم المحافظة باللغة العربية', 'اسم المنطقة باللغة العربية']]

#     # Apply the helper function to each row of the target DataFrame
#     closest_districts_info = target_df.apply(
#         lambda row: get_closest_district_info(row['longitude'], row['latitude']),
#         axis=1
#     )

#     # Concatenate the closest districts information to the target DataFrame
#     target_df = pd.concat([target_df, closest_districts_info], axis=1)

#     return target_df

import pandas as pd
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculates the Haversine distance between two points on Earth.

    Args:
        lat1 (float): Latitude of the first point.
        lon1 (float): Longitude of the first point.
        lat2 (float): Latitude of the second point.
        lon2 (float): Longitude of the second point.

    Returns:
        float: The distance in kilometers.
    """
    R = 6371  # Radius of Earth in kilometers

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance


def add_closest_district_info_to_target_df(districts, target_df):
    """
    Adds the 'name', 'اسم المحافظة باللغة العربية', and 'اسم المنطقة باللغة العربية'
    columns of the closest district to each row in the target DataFrame.

    Args:
        districts (pd.DataFrame): DataFrame with district information
                                  ('name', 'longitude', 'latitude',
                                   'اسم المحافظة باللغة العربية',
                                   'اسم المنطقة باللغة العربية').
        target_df (pd.DataFrame): DataFrame to which the closest district
                                 information will be added ('longitude', 'latitude').

    Returns:
        pd.DataFrame: The target_df with the added columns.
    """

    def get_closest_district_info(target_lon, target_lat):
        """Helper function to find the closest district for a single target coordinate
           and return the specified columns.
        """
        distances = districts.apply(
            lambda row: haversine(target_lat, target_lon, row['latitude'], row['longitude']),
            axis=1
        )
        closest_district_index = distances.idxmin()  # Index of the minimum distance
        closest_district = districts.loc[closest_district_index]
        return closest_district[['name', 'اسم المحافظة باللغة العربية', 'اسم المنطقة باللغة العربية']]

    # Apply the helper function to each row of the target DataFrame
    def apply_closest_district_info(row):
        if pd.isna(row['longitude']) or pd.isna(row['latitude']):
            return pd.Series([None, None, None], index=['name', 'اسم المحافظة باللغة العربية', 'اسم المنطقة باللغة العربية'])
        else:
            return get_closest_district_info(row['longitude'], row['latitude'])

    closest_districts_info = target_df.apply(apply_closest_district_info, axis=1)

    # Concatenate the closest districts information to the target DataFrame
    target_df = pd.concat([target_df, closest_districts_info], axis=1)

    return target_df
```

```python
import pandas as pd

def extract_date_time(DataFrame, Start_Date_Time, Start_Date, Start_Time):
    """
    تقوم هذه الدالة باستخراج التاريخ والوقت من عمود datetime في DataFrame.

    Args:
        DataFrame (pd.DataFrame): DataFrame المدخل.
        Start_Date_Time (str): اسم العمود الذي يحتوي على بيانات التاريخ والوقت.
        Start_Date (str): اسم العمود الجديد للتاريخ.
        Start_Time (str): اسم العمود الجديد للوقت.

    Returns:
        pd.DataFrame: DataFrame مع الأعمدة الجديدة للتاريخ والوقت.
    """

    # تحويل نوع العمود
    DataFrame[Start_Date_Time] = pd.to_datetime(DataFrame[Start_Date_Time])

    # استخراج الأعمدة

    ## التاريخ
    DataFrame[Start_Date] = DataFrame[Start_Date_Time].dt.date
    DataFrame[Start_Date] = pd.to_datetime(DataFrame[Start_Date], format='%Y-%m-%d')

    ## الوقت
    DataFrame[Start_Time]  = DataFrame[Start_Date_Time].dt.time
    DataFrame[Start_Time]  = pd.to_datetime(DataFrame[Start_Time] ,format='%H:%M:%S').dt.time

    # إرجاع DataFrame الجديد
    return DataFrame
```

```python
def Weekly_Complaints(df, start_date_main, end_date_main, date_column='Complaint_Start_Date'):
    """
    Calculates the number of complaints within a main date range and for preceding weeks.

    Args:
        df (pd.DataFrame): The DataFrame containing complaint data.
        start_date_main (str or pd.Timestamp): The start date of the main range (YYYY-MM-DD).
        end_date_main (str or pd.Timestamp): The end date of the main range (YYYY-MM-DD).
        date_column (str, optional): The name of the date column. Defaults to 'Complaint_Start_Date'.

    Returns:
        None: Prints the number of complaints for the main range and preceding weeks.
    """

    # Convert date strings to datetime objects
    df[date_column] = pd.to_datetime(df[date_column])
    start_date_main = pd.to_datetime(start_date_main)
    end_date_main = pd.to_datetime(end_date_main)

    # Calculate visits in the main date range
    main_range_visits = df[(df[date_column] >= start_date_main) & (df[date_column] <= end_date_main)].shape[0]
    print(f"Complaints from {start_date_main.strftime('%Y-%m-%d')} to {end_date_main.strftime('%Y-%m-%d')}: {main_range_visits}")

    # Calculate visits for each preceding week
    current_start = start_date_main - pd.Timedelta(days=7)
    current_end = start_date_main - pd.Timedelta(days=1)

    while current_start >= df[date_column].min():
        weekly_visits = df[(df[date_column] >= current_start) & (df[date_column] <= current_end)].shape[0]
        print(f"Complaints from {current_start.strftime('%Y-%m-%d')} to {current_end.strftime('%Y-%m-%d')}: {weekly_visits}")

        # Move to the previous week
        current_end = current_start - pd.Timedelta(days=1)
        current_start = current_start - pd.Timedelta(days=7)
```
